### Result

# Total Price Validation

The TotalPrice column was validated by comparing it against Quantity × UnitPrice calculations.

```python
(df['Quantity'] * df['UnitPrice']) == df['TotalPrice']
```

```python
((df['Quantity'] * df['UnitPrice']) != df['TotalPrice']).sum()
```

```python
invalid_totals = df[
    (df['Quantity'] * df['UnitPrice']) != df['TotalPrice']
]

invalid_totals.head()
```

```python
invalid_totals.shape
```

```python
df['CalculatedTotal'] = df['Quantity'] * df['UnitPrice']
```

```python
df[['Quantity', 'UnitPrice', 'TotalPrice', 'CalculatedTotal']].head(10)
```

```python
(
    round(df['Quantity'] * df['UnitPrice'], 2)
    != round(df['TotalPrice'], 2)
).sum()
```

Initial validation checks suggested inconsistencies between Quantity × UnitPrice and TotalPrice values. However, after applying currency rounding to two decimal places, all calculations were confirmed to be accurate. The discrepancy was caused by floating-point precision behavior in Python.

# Text Standardization

Text fields were standardized to improve consistency and readability across the dataset.

```python
text_columns = [
    'Product',
    'PaymentMethod',
    'OrderStatus',
    'CouponCode',
    'ReferralSource'
]

for col in text_columns:
    df[col] = df[col].str.title()
```
